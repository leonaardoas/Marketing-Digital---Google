```python
df = pd.read_csv("/content/drive/MyDrive/Projetos Machine Learning /Marketing Digital/train.csv",
                 dtype = {'date':object, 'fullVisitorId':object, 'visitId':object})
```

```python
df.dtypes
```

```python
import json
```

```python
dic = ['device', 'geoNetwork', 'trafficSource', 'totals']
for coluna in dic:
  df = df.join(pd.DataFrame([json.loads(linha) for linha in df[coluna]]))
```

```python
df
```

```python
df.drop(dic, axis=1, inplace=True)
```

```python
df
```

```python
df.drop('adwordsClickInfo', axis=1, inplace=True)
```

```python
df
```

```python
coluna_na = []
for coluna in df.columns:
    print(coluna + ':' + str(df[coluna].nunique()))
    if df[coluna].nunique() ==1:
      coluna_na.append(coluna)
```

```python
coluna_na
```

```python
df.drop(coluna_na, axis=1, inplace=True)
```

```python
df
```

```python
quant = ['bounces', 'hits', 'newVisits', 'pageviews', 'transactionRevenue']

for coluna in quant:
  df[coluna] = pd.to_numeric(df[coluna])
```
